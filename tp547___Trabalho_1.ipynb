4) Distribuição Geométrica

Alínea a) R:

A simulação baseia-se em experimentos de Bernoulli independentes. De acordo com o enunciado, a variável aleatória $Y$ segue a distribuição:
$$P(Y = n) = (1 - p)^n p

```
import random

def simular_geometrica(p):
    n_insucessos = 0
    while True:
        # random.random() gera um valor entre 0 e 1
        if random.random() <= p:
            return n_insucessos
        n_insucessos += 1

# Gerando uma amostra de 10 resultados
p = 0.3
amostras = [simular_geometrica(p) for _ in range(10)]
print(f"Amostras geradas (nº de insucessos): {amostras}")
```

5. Método da Inversa

Alíne d) R:

```
import numpy as np
import matplotlib.pyplot as plt

u = np.random.uniform(0, 1, 1000)

# transformação inversa x = sqrt(u)
amostras = np.sqrt(u)

# Plotar o histograma e a PDF teórica
plt.hist(amostras, bins=30, density=True, alpha=0.6, color='skyblue', label='Amostras Simuladas')
x = np.linspace(0, 1, 100)
plt.plot(x, 2*x, 'r-', lw=2, label='PDF Teórica $f(x)=2x$')
plt.legend()
plt.show()
```
